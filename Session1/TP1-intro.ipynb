Plot of data
===

All imports and settings
---

```python
import os
import numpy as np
from scipy.io import loadmat

import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

data1_path = os.path.join(os.curdir, 'data1')
data2_path = os.path.join(os.curdir, 'data2')
```

Plot for data 1
---

```python
data_fnames = os.listdir(data1_path)
data_fnames.sort()
fig, *axis = plt.subplots(nrows=2, ncols=2, figsize=(15,10))
axis = np.array(axis).flatten()
fig.suptitle("data1")
for ax, fname in zip(axis, data_fnames):
    mat = loadmat(os.path.join(data1_path, fname))
    ax.set_title(fname.split('.')[0])
    ax.scatter(mat['X'][:,0], mat['X'][:,1])
```
